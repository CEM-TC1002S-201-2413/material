# Visualizaciones

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/vgsales.csv")
```

```python
df.head()
```

```python
numerical_columns = [col for col in df.columns if df[col].dtype != "O"]
correlation_matrix = df[numerical_columns].corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True)
```

```python
sns.pairplot(df[numerical_columns], 
             diag_kind="kde",
             kind="reg",
             corner=True,
             plot_kws={
                 "line_kws": {
                     "color": "#fa165b"
                     },
                 "scatter_kws": {
                     "alpha": 0.5
                 }
                 }
             )
```

## Gráficas de barras

```python
df.plot.bar(x="Genre", y="Global_Sales")
```

```python
df.groupby("Genre")["Global_Sales"].sum().plot.bar()
```

```python
df.groupby(["Publisher", "Platform"])[["NA_Sales", "Global_Sales"]].mean()
```

```python
sns.barplot(data=df, 
            x="Genre", 
            y="Global_Sales", 
            hue="Genre",
            estimator="sum")
plt.xticks(rotation=90)
plt.show()
```

## Gráfica de líneas

```python
df.groupby("Year")["Global_Sales"].mean().plot.line()
```

```python
sns.lineplot(data=df, x="Year", y="Global_Sales", estimator="mean")
```

## Gráficas de dispersión

```python
df.plot.scatter(x="NA_Sales", y="Global_Sales", alpha=0.5)
```

```python
sns.scatterplot(data=df, 
                x="NA_Sales", 
                y="Global_Sales", 
                hue="Genre",
                alpha=0.5)
```

## Histogramas

```python
df["Year"].plot.hist(bins=10)
```

```python
sns.histplot(data=df, x="Year", bins=10)
```

## Boxplot (cajas y bigotes)

```python
df["Year"].plot.box()
```

```python
sns.boxplot(data=df, y="Year")
```

## Gráficas de pie/pastel

```python
df["Genre"].value_counts().plot.pie(autopct="%2.1f%%")
```

## KDE

```python
df["Year"].plot.kde()
```

```python
sns.kdeplot(data=df, x="Year")
```
